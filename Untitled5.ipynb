```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data_train = pd.read_csv(r'/content/train.csv')
test_data = pd.read_csv(r'/content/test.csv')
```

```
import statsmodels.tsa.stattools as sts
import statsmodels.graphics.tsaplots as sgt

from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.tsa.arima_model import ARMA,ARIMA
from scipy.stats.distributions import chi2

from math import sqrt
from pmdarima.arima import auto_arima
#from arch import arch_model
```

```
def predict(train,test):
    
    model_ar = auto_arima(train,m=5)
    
    start_date = str(test.index[0])
    end_date = str(test.index[-1])
    test_auto_pred = pd.DataFrame(model_ar.predict(n_periods=len(test[start_date:end_date])),
                              index=test[start_date:end_date].index)
    return test_auto_pred
```

```
data_jun_1 = data_train[data_train['Junction']==1]
data_jun_1.asfreq('h').isna().sum()

data_jun_1.drop('Junction',axis=1,inplace=True)
size = int(len(data_jun_1)*0.8)
train = data_jun_1.iloc[:size]
test_jun_1 = test_data[test_data['Junction']==1]
test_jun_1.asfreq('h').isna().sum()

pred_1 = predict(train,test_jun_1)
```

```
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
from keras.preprocessing.sequence import TimeseriesGenerator

from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
```

```
size = int(len(data_jun_1)*0.9)
train1,test1 = data_jun_1.iloc[:size],data_jun_1.iloc[size:]

scaler = MinMaxScaler()
scaler.fit(train1)
train1 = scaler.transform(train1)
test1 = scaler.transform(test1)

n_input = 24 ##cycle
n_features = 1  ##vehicles

generator = TimeseriesGenerator(data=train1,targets=train1,length=n_input,batch_size=6)
```

```
model = Sequential()

model.add(LSTM(200,activation='relu',input_shape=(n_input,n_features)))
model.add(Dropout(0.15))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')

model.fit_generator(generator,epochs=5)
```

```
pred_list = []

batch = train1[-n_input:].reshape((1,n_input,n_features))
for i in range(n_input):
    pred_list.append(model.predict(batch)[0])
    batch = np.append(batch[:,1:,:],[[pred_list[i]]],axis=1)
```

```
df_prediction = pd.DataFrame(scaler.inverse_transform(pred_list),index=data_jun_1[-n_input:].index,columns=['predictions'])

df_test = pd.concat([data_jun_1,df_prediction],axis=1)
```
